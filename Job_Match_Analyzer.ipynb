# **Importing Libraries**

```python
import re
import pdfplumber
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# **Upload CV**


```python
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
```

# **Cleaning text in CV**


```python
cv_text = ""

with pdfplumber.open(file_name) as pdf:
    for page in pdf.pages:
        cv_text += page.extract_text() + " "
```

```python
def fix_spaced_text(text):
    words = text.split()
    fixed_text = ""
    temp_word = ""

    for word in words:
        if len(word) == 1:
            temp_word += word
        else:
            if temp_word:
                fixed_text += temp_word + " "
                temp_word = ""
            fixed_text += word + " "

    if temp_word:
        fixed_text += temp_word

    return fixed_text

cv_text = fix_spaced_text(cv_text)
```

# **Uploading job description**

```python
job_description = input("Paste Job Description here: ")
```

# **Cleaning text in job description**


```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

cv_clean = clean_text(cv_text)
job_clean = clean_text(job_description)
```

# **Similarity** **Calculation**

```python
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
vectors = vectorizer.fit_transform([cv_clean, job_clean])

similarity = cosine_similarity(vectors[0], vectors[1])
score = similarity[0][0] * 100
```

# **Skill bank list**

```python
skill_bank = [
    "python",
    "r",
    "sql",
    "excel",
    "power bi",
    "looker studio",
    "tableau",
    "data analysis",
    "data visualization",
    "machine learning",
    "statistics",
    "communication",
    "problem solving",
    "critical thinking",
    "teamwork",
    "pandas",
    "numpy",
    "scikit learn",
    "scikit-learn",
    "flutter",
    "java",
    "c++",
    "data structures",
    "artificial intelligence",
    "analytical skills",
    "attention to detail"
]
```

# **Extract required skills from the job description**

```python
required_skills = []

for skill in skill_bank:
    if skill in job_clean:
        required_skills.append(skill)
```

# **Compare required skills with cv**

```python
matched_skills = []
missing_skills = []

for skill in required_skills:
    if skill in cv_clean:
        matched_skills.append(skill)
    else:
        missing_skills.append(skill)

text_score = score

if required_skills:
    skill_score = (len(matched_skills) / len(required_skills)) * 100
else:
    skill_score = 0
```

# **Analysis output**

```python
print("===== JOB MATCH ANALYSIS =====")
print(f"Text Similarity Score: {round(text_score, 2)} %")
print(f"Skill Match Score: {round(skill_score, 2)} %")

print("\nMatched Skills:")
for skill in matched_skills:
    print("-",skill)

print("\nMissing Skills:")
for skill in missing_skills:
    print("-",skill)
```
